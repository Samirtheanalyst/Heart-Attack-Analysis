```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\Samir\\OneDrive\\Desktop\\Dataset\\heart_attack_prediction_india.csv")
```

```python
# 1. Total Patients
total_patients = len(df)
print(f"1. Total Patients: {total_patients}\n")
```

```python
# 2. Average age of patients from each state
avg_age_by_state = df.groupby('State_Name')['Age'].mean().reset_index().sort_values('Age', ascending=False)
print("2. Average Age by State:")
print(avg_age_by_state)
```

```python

# 2. Plot: Average Age by State
plt.figure(figsize=(12, 6))
plt.barh(avg_age_by_state['State_Name'], avg_age_by_state['Age'], color='skyblue')
plt.xlabel('Average Age')
plt.title('Average Age by State')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 3. Number of male and female patients
gender_count = df['Gender'].value_counts()
print("\n3. Gender Distribution:")
print(gender_count)
```

```python
# 3. Plot: Gender Distribution
gender_count.plot(kind='bar', color=['lightcoral', 'lightblue'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Top 5 states with the highest average cholesterol levels
top5_cholesterol_states = df.groupby('State_Name')['Cholesterol_Level'].mean().nlargest(5).reset_index()
print("\n4. Top 5 States by Average Cholesterol Level:")
print(top5_cholesterol_states)
```

```python
# 4. Plot: Top 5 States by Average Cholesterol Level
plt.figure(figsize=(8, 5))
plt.barh(top5_cholesterol_states['State_Name'], top5_cholesterol_states['Cholesterol_Level'], color='goldenrod')
plt.xlabel('Average Cholesterol Level')
plt.title('Top 5 States by Cholesterol Level')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 5. Patients with diabetes, hypertension, and obesity
count_with_conditions = df[(df['Diabetes'] == 1) & (df['Hypertension'] == 1) & (df['Obesity'] == 1)].shape[0]
print(f"\n5. Patients with Diabetes, Hypertension, and Obesity: {count_with_conditions}")
```

```python
# 6. Average heart attack risk for smokers vs non-smokers
smoking_risk = df.groupby('Smoking')['Heart_Attack_Risk'].mean().reset_index()
print("\n6. Average Heart Attack Risk (Smoking vs Non-Smoking):")
print(smoking_risk)
```

```python
# 6. Plot: Heart Attack Risk for Smokers vs Non-Smokers
sns.barplot(x='Smoking', y='Heart_Attack_Risk', data=smoking_risk)
plt.xticks([0, 1], ['Non-Smoker', 'Smoker'])
plt.title('Heart Attack Risk: Smokers vs Non-Smokers')
plt.ylabel('Average Heart Attack Risk')
plt.show()
```

```python
# 7. Average stress level by gender
stress_by_gender = df.groupby('Gender')['Stress_Level'].mean().reset_index()
print("\n7. Average Stress Level by Gender:")
print(stress_by_gender)
```

```python
# 7. Plot: Average Stress Level by Gender
sns.barplot(x='Gender', y='Stress_Level', data=stress_by_gender)
plt.title('Average Stress Level by Gender')
plt.show()
```

```python
# 8. State with the highest average heart attack risk
highest_risk_state = df.groupby('State_Name')['Heart_Attack_Risk'].mean().idxmax()
print(f"\n8. State with Highest Heart Attack Risk: {highest_risk_state}")
```

```python
# 9. Average annual income of patients with a history of heart attacks
avg_income_heart_attack_patients = df[df['Heart_Attack_History'] == 1]['Annual_Income'].mean()
print(f"\n9. Average Annual Income (Heart Attack History): {avg_income_heart_attack_patients:.2f}")
```
